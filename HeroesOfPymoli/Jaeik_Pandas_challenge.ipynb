```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data["SN"].count()

player_count = pd.DataFrame({"Total Players" : [total_players]})

player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data ["Item Name"].unique())

# print(unique_item)

average = purchase_data["Price"].mean()

# print(average)

number_purchase = purchase_data["SN"].count()

# print(number_purchase)

total_revenue = purchase_data["Price"].sum()

# print(total_revenue)

summary = pd.DataFrame({"Number of Unique Items" : [unique_item],
                       "Average Price" : [average],
                        "Number of Purchases" :[number_purchase],
                        "Total Revenue" : [total_revenue]
                       })

summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:.2f}".format)


summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = purchase_data["Gender"].value_counts()

gender_count_df = pd.DataFrame(gender_count)

# gender_count_df

gender_total = gender_count_df["Gender"].sum()

Percentage_Players = gender_count_df["Gender"]/gender_total

gender_count_df["Percentage of Players"] = Percentage_Players

# gender_count_df

gender_count_df = gender_count_df.rename(columns={"Gender" : "Total Count"})

gender_count_df["Percentage of Players"] = gender_count_df["Percentage of Players"].map("{:.2%}".format)

gender_count_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_count = purchase_data["Gender"].value_counts()

# gender_count_df = pd.DataFrame(gender_count)

# gender_count_df

group_df = purchase_data.groupby("Gender")

average = group_df["Price"].mean()

total = group_df["Price"].sum()

Per_person = total/gender_count

Consolidation_df = pd.DataFrame({"Total count" : gender_count,
                                "Average Purchase Price" : average,
                                 "Total Purchase Value" : total,
                                 "Avg Total Purchase per Person" : Per_person
                                })

Consolidation_df["Average Purchase Price"] = Consolidation_df["Average Purchase Price"].map("${:.2f}".format)
Consolidation_df["Total Purchase Value"] = Consolidation_df["Total Purchase Value"].map("${:.2f}".format)
Consolidation_df["Avg Total Purchase per Person"] = Consolidation_df["Avg Total Purchase per Person"].map("${:.2f}".format)

# change row order

Consolidation_df = Consolidation_df.reindex(["Female","Male","Other / Non-Disclosed"])

Consolidation_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,13.9,18.9,23.9,28.9,33.9,38.9,100]

group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
               
purchase_data ["Age category"] = pd.cut(purchase_data["Age"],bins, labels=group_names, include_lowest=True)

# purchase_data

age_count = purchase_data["Age category"].value_counts()

age_count_df = pd.DataFrame(age_count)

age_count_df = age_count_df.reindex(["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"])

age_count_total = age_count_df["Age category"].sum()

Percentage = age_count_df["Age category"]/age_count_total

age_count_df["Percentage of Players"] = Percentage

age_count_df["Percentage of Players"] = age_count_df["Percentage of Players"].map("{:.2%}".format)

age_count_df = age_count_df.rename(columns={"Age category" : "Total Count"})

age_count_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# initiating df and renaming column names

aggregated_purchase_data_items = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

# conducting 3 groupbys on: price, count, and total

aggregated_purchase_data_items_price = aggregated_purchase_data_items.groupby(["Item ID", "Item Name"]).max().rename(columns = {'Price': 'Item Price'})

aggregated_purchase_data_items_grpby_itemCount = aggregated_purchase_data_items.groupby(["Item ID", "Item Name"]).count().rename(columns = {'Price': 'Purchase Count'})

aggregated_purchase_data_items_grpby_itemSum = aggregated_purchase_data_items.groupby(["Item ID", "Item Name"]).sum().rename(columns = {'Price': 'Total Purchase Value'})

# Combining tables by using join / ascending order

combined_groupby_items_popular = aggregated_purchase_data_items_grpby_itemCount.join(aggregated_purchase_data_items_price).join(aggregated_purchase_data_items_grpby_itemSum).sort_values(by = ['Purchase Count'], ascending=False)

# Reformatting

combined_groupby_items_popular['Item Price'] = combined_groupby_items_popular['Item Price'].map("${:.2f}".format)

combined_groupby_items_popular['Total Purchase Value'] = combined_groupby_items_popular['Total Purchase Value'].map("${:.2f}".format)

# result
combined_groupby_items_popular.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# identifying the most profitable items by sorting by 'Total Purchase Value' after joining dataframes obtained from groupby

combined_groupby_items_profitable = aggregated_purchase_data_items_grpby_itemCount.join(aggregated_purchase_data_items_price).join(aggregated_purchase_data_items_grpby_itemSum).sort_values(by = ['Total Purchase Value'], ascending=False)

# Reformatting

combined_groupby_items_profitable['Item Price'] = combined_groupby_items_profitable['Item Price'].map("${:.2f}".format)

combined_groupby_items_profitable['Total Purchase Value'] = combined_groupby_items_profitable['Total Purchase Value'].map("${:.2f}".format)

# result

combined_groupby_items_profitable.head()
```
